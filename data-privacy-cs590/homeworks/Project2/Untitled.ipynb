```julia
using Plots
```

```julia
# reading file into matrix
M = readdlm("programming-task/adult.data",',')
```

```julia
AgeLessDict = Dict{Int32, Int32}()
AgeHigherDict = Dict{Any, Int32}()
workClassHigher= Dict{Any, Int32}()
workClassLess= Dict{Any, Int32}()
raceHigher= Dict{Any, Int32}()
raceLess= Dict{Any, Int32}()
```

```julia
sort(unique(M[:,11]))
```

```julia
#race
for i in M[:,9]
    raceHigher[i]=0
    raceLess[i]=0
end
j = 1
for i in M[:,9]
    if (M[j,15] ==" <=50K")
        raceLess[i] = raceLess[i]+1
    else
        raceHigher[i] = raceHigher[i]+1
    end
    j = j+1
end
a = sort(collect(keys(raceLess)))
b = []
for i in a 
   push!(b,raceLess[i])
end
scatter(b)
a = sort(collect(keys(raceHigher)))
b = []
for i in a 
   push!(b,raceHigher[i])
end
scatter!(b)
```

```julia
#age
for i in M[:,1]
    AgeLessDict[i]=0
    AgeHigherDict[i]=0
end
j = 1
for i in M[:,1]
    if (M[j,15] ==" <=50K")
        AgeLessDict[i] = AgeLessDict[i]+1
    else
        AgeHigherDict[i] = AgeHigherDict[i]+1
    end
    j = j+1
end
a = sort(collect(keys(AgeHigherDict)))
b = []
for i in a 
   push!(b,AgeHigherDict[i])
end
scatter(b)
a = sort(collect(keys(AgeLessDict)))
b = []
for i in a 
   push!(b,AgeLessDict[i])
end
scatter!(b)
```

```julia
#working class
for i in M[:,2]
    workClassHigher[i]=0
    workClassLess[i]=0
end
j = 1
for i in M[:,2]
    if (M[j,15] ==" <=50K")
        workClassLess[i] = workClassLess[i]+1
    else
        workClassHigher[i] = workClassHigher[i]+1
    end
    j = j+1
end
a = sort(collect(keys(workClassHigher)))
b = []
for i in a 
   push!(b,workClassHigher[i])
end
plot(b)
```
